## 3b. Missing Values

```python
# Checking for missing values
missing_values=df.isnull().sum()

# Calculate the total number of rows in the DataFrame
total_cells = df.shape[0]

# Calculate the percentage of missing values
percent_missing = (missing_values / total_cells) * 100

# Round the decimal to 2 decimal places
percent_missing = percent_missing.round(2)

# Convert the float type to string
percent_missing = percent_missing.astype(str)

# Add the '%' symbol to each value
percent_missing = percent_missing + '%'

# Combine the two Series into a new DataFrame
missing_value_table = pd.concat([missing_values, percent_missing], axis=1)

# Rename the columns of the DataFrame
missing_value_table.columns = ['Missing Values', 'Percent Missing']

# Sort the resulting DataFrame based on the 'Percent Missing' column in ascending order
missing_value_table.sort_values('Percent Missing', ascending=True, inplace=True)

# Print the resulting table
print(missing_value_table)
```

Observations: According to the Description of the dataset document, the lea_beds column has the same info as the aggeration_code colum. It uniquely identifies the entity (school, district, etc.) and may be useful to researchers who want to select or aggregate data for specific entities. I'm going to ignore the missing values and drop this column.

The column Lea_name is also not necessary for this analysis, hence it will be dropped.

nrc_code and nrc_desc has 8164 missing values in each column, which is approximately about 3.6% which is relatively low compared to the overall dataset size. Therefore I will ignore these missing values in the nrc_desc but dropping nrc_code as it's a unwanted column.

county_code, county_name have 1098 missing value in each columnm, which is only about 0.45%. It is relatively low compared to the overall dataset size. Therefore I will ignore these missing values.

boces_code, boces_name has 9120 missing values. I'm not sure if I will need to use these columns so I'm going to   ignore these missing values in the boces_name but dropping boces_code as it's a unwanted column.
